```python
# function to count the number of lines in a file
def file_len(file_name):
  """ Count number of lines in a file."""
  f = open(file_name)
  nr_of_lines = sum(1 for line in f)
  f.close()
  return nr_of_lines-1 # excluding the first line
```

```python
# print the number of lines in each file and add them to a list to compute statistics
l1=[]
for file in fhvhv:
    print("The number of lines in ",file,"is ",file_len(file))
    l1.append(file_len(file))

l2=[]
for file in fhv:
    print("The number of lines in ",file,"is ",file_len(file))
    l2.append(file_len(file))

l3=[]
for file in green:
    print("The number of lines in ",file,"is ",file_len(file))
    l3.append(file_len(file))

l4=[]
for file in yellow:
    print("The number of lines in ",file,"is ",file_len(file))
    l4.append(file_len(file))
           
        
```
